```python
import pandas as pd
df = pd.read_csv('laptops.csv', sep=';')
df.head()
```

```python
df['Cpu_Company']=df['Cpu'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
df['Cpu_Company'].value_counts()
```

```python
def convert_to_gb(x):
    memory=x.split(' ')[0]
    if memory.endswith('GB'):
        res=memory.replace('GB','')
    elif memory.endswith('TB'):
        res=float(memory.replace('TB',''))*1024
    return int(res)
df['Memory_Amount']=df['Memory'].apply(convert_to_gb)
df.head()
```

```python
df['Memory'].value_counts()
```

```python
def Type(x):
    if '+' in x:
        mem1=x.split('B')[1][1:x.split('B')[1].find('+')-1]
        mem2=x.split('B')[-1][1:]
        res=mem1+ ' ' +mem2
    else: res=mem2=x.split('B')[-1][1:]
    return res
df['Memory_Type']=df['Memory'].apply(Type)
df.head()
```

```python
df['Memory_Type'].value_counts()
```

```python
df.drop(columns=['Memory','ScreenResolution'],inplace=True)
df
```

```python
df['SSD']=0
df.head()
```
